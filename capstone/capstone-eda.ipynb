```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# what types of warnings for this notebook
import warnings
warnings.filterwarnings('ignore')

# create initial dataframe
items = pd.read_csv('items.csv')

items.head()
```

```python
items.describe()
```

```python
items.info()
```

```python
# data to be plotted
prices = items['Price'].fillna(items['Price'].mean())

fig = plt.figure(figsize=(30,20))
ax = fig.add_subplot(111)

sns.distplot(prices, hist = False, kde = True, rug=True,
                 kde_kws = {'linewidth': 3, 'shade': True, 'color': "green"},
                 rug_kws={"color": "black", "alpha":0.3, "linewidth": 2, "height":0.1},
                 label = "price")

# labels
plt.tick_params(axis='both',color='white',labelcolor='white')
plt.title("Price Distribution",color="white")
plt.xlabel("price",color='white')
plt.ylabel("density",color='white')
plt.show()
```
