```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# create initial dataframe
items_df = pd.read_csv('items.csv')

items_df.head()
```

```python
from scipy import stats
import matplotlib.mlab as mlab
import numpy as np

# data to be plotted
prices = items_df['Price'].fillna(items_df['Price'].mean())

# plot histogram
n, bins, patches = plt.hist(prices, bins=80, density=True)

# get scope of distribution
xmin, xmax = prices.min(), prices.max()
x = np.linspace(xmin, xmax)

# get mean and standard deviation  
m, s = stats.norm.fit(prices) 

# get y values for PDF
y = stats.norm.pdf(x, m, s)

# plot fit line
plt.plot(x, y, 'r--')

# labels
plt.tick_params(axis='both',color='white',labelcolor='white')
plt.title("Price Distribution",color="white")
plt.xlabel("price",color='white')
plt.ylabel("quantity",color='white')
plt.show()

## TODO: need to find a better fit line for the distribution
```
